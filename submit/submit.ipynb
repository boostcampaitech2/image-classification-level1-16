```python
import numpy as np
import pandas as pd
```

```python
# 레이블된 csv 파일 불러오기

df_submit = pd.read_csv('submit_labeled.csv')
df_submit.head(5)

# gender) male:0, female:1
# mask) wear:0, incorrect:1, normal:2
# age) ~30:0, 30~59:1, 60~:2
```

```python
# 이 dataframe에 자기가 예측한 레이블을 원하는 항목에 넣으세요

# 예시

my_pred = [0]*len(df_submit)
df_submit['mask'] = my_pred
df_submit.head()
```

```python
# 이 3개의 레이블을 취합합니다.

df_submit['ans'] = df_submit['age'] + 3 * df_submit['gender'] + 6 * df_submit['mask']
df_submit.head()
```

```python
# 'ImageID', 'ans' 만 뽑아서 현재 디렉토리에 저장합니다.

df_submit[['ImageID', 'ans']].to_csv('submit_example.csv', index=False)
```
