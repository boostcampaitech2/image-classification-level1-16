```python
#빈도수 고려
add_dict={}
for key, value in item_dict.items():
    value_set=set(value)
    value_li=list(value)
    count_dict={}
    for a in value_li:
        c=value.count(a)
        count_dict[a]=c
    count_dict_sort=sorted(count_dict.items(), reverse=True, key=lambda item:item[1])
    add_dict[key]=count_dict_sort
add_dict
```

top 5/10/15/20

```python
top_5_dict={}
top_10_dict={}
top_15_dict={}
top_20_dict={}
for key, value in add_dict.items():
    top_5=[]
    top_10=[]
    top_15=[]
    top_20=[]
    for items in value[:5]:
        top_5.append(items[0])
    for items in value[:10]:
        top_10.append(items[0])
    for items in value[:15]:
        top_15.append(items[0])
    for items in value[:20]:
        top_20.append(items[0])
    top_5_dict[key]=top_5
    top_10_dict[key]=top_10
    top_15_dict[key]=top_15
    top_20_dict[key]=top_20
# 빈도수 튜플로 저장된 딕셔너리 형태
#     top_5_dict[key]=value[:5]
#     top_10_dict[key]=value[:10]
#     top_15_dict[key]=value[:15]
#     top_20_dict[key]=value[:20]

top_5_dict
```

```python
count_top_5=0
count_top_10=0
count_top_15=0
count_top_20=0
for i, row in new.iterrows():
    pd_c=row['pd_c']
    top_5=row['top_5']
    top_10=row['top_10']
    top_15=row['top_15']
    top_20=row['top_20']
    count_5=0
    count_10=0
    count_15=0
    count_20=0
    for a in pd_c:
        if a in top_5:
            count_5+=1
        if a in top_10:
            count_10+=1
        if a in top_15:
            count_15+=1
        if a in top_20:
            count_20+=1
    if count_5>1:
        count_top_5+=1
    if count_10>1:
        count_top_10+=1
    if count_15>1:
        count_top_15+=1
    if count_20>1:
        count_top_20+=1
print(count_top_5)
print(count_top_10)
print(count_top_15)
print(count_top_20)
```

버전 2 : 빈도를 순위로 바꿈

```python
import pandas as pd
#2. 상품 빈도수 가중치
top_5_score={}
top_10_score={}
top_15_score={}
top_20_score={}

score_5={}
for key, value in add_dict.items():
    top_5=[]
    top_10=[]
    top_15=[]
    top_20=[]
    for items in value[:5]:
        top_5.append(items[1])
    for items in value[:10]:
        top_10.append(items[1])
    for items in value[:15]:
        top_15.append(items[1])
    for items in value[:20]:
        top_20.append(items[1])
        
    top_5=pd.Series(top_5)
    top_10=pd.Series(top_10)
    top_15=pd.Series(top_15)
    top_20=pd.Series(top_20)
    
    rank_5=top_5.rank(method='min', ascending=False)
    rank_10=top_10.rank(method='min', ascending=False)
    rank_15=top_15.rank(method='min', ascending=False)
    rank_20=top_20.rank(method='min', ascending=False)
    
    rank_5=list(rank_5)
    rank_10=list(rank_10)
    rank_15=list(rank_15)
    rank_20=list(rank_20)
        
    top_5_score[key]=rank_5
    top_10_score[key]=rank_10
    top_15_score[key]=rank_15
    top_20_score[key]=rank_20

    
top_10_score
```

```python
score_df=new
score_df['top_5_score']=list(top_5_score.values())#왜인지 모르겠는데 top_5만 튜플로 반환됨 - 리스트로 바꿔줌
score_df['top_10_score']=top_10_score.values()
score_df['top_15_score']=top_15_score.values()
score_df['top_20_score']=top_20_score.values()
score_df
```

```python
count_5=0
count_10=0
count_15=0
count_20=0
for i, row in score_df.iterrows():
    pd_c=row['pd_c']
    top_5=row['top_5']
    top_10=row['top_10']
    top_15=row['top_15']
    top_20=row['top_20']
    
    top_5_score=row['top_5_score']
    top_10_score=row['top_10_score']
    top_15_score=row['top_15_score']
    top_20_score=row['top_20_score']

    for a in pd_c:
        if a in top_5:
            num=top_5.index(a)
            score=top_5_score[num]
            count_5+=1/score
        if a in top_10:
            num=top_10.index(a)
            score=top_10_score[num]
            count_10+=1/score
        if a in top_15:
            num=top_15.index(a)
            score=top_15_score[num]
            count_15+=1/score
        if a in top_20:
            num=top_20.index(a)
            score=top_20_score[num]
            count_20+=1/score
print(count_5)
print(count_10)
print(count_15)
print(count_20)
```

```python
top_5_acc=count_5/len(score_df)
top_10_acc=count_10/len(score_df)
top_15_acc=count_15/len(score_df)
top_20_acc=count_20/len(score_df)
print(top_5_acc,top_10_acc,top_15_acc,top_20_acc)
```
